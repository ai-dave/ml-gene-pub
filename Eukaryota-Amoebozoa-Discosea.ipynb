```python
kv_ncbi = kvec.KmerVectors(['A','G','C','T'], 6, ncbidatasets=[flabellinia_nds,longamoebia_nds], verbose=True)
print(kv_ncbi.labels)
```

```python
d_ncbi = kv_ncbi.seq2KmerSentences(seq_type='fna2', base_count_max=4, length_min=500, dataset_limit=1300)
```
